# Data Visualization Exercise 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Read the stock prices data using pandas
stock_df = pd.read_csv('stock_data.csv')
stock_df.head()
```

```python
print(plt.style.available)
plt.style.use('fivethirtyeight')
```

```python
stock_df.plot(x = 'Date', y = 'FB', label = 'Facebook Stock Price', figsize = (15,10), linewidth = 2)
plt.ylabel('Price ($)', color = 'red')
plt.xlabel('')
plt.title('Facebook Stock Graph', color = 'red');
```

# PLOT SCATTERPLOT 

```python
# Read daily return data using pandas

daily_return_df = pd.read_csv('stocks_daily_returns.csv')
daily_return_df.head()
```

```python
x = daily_return_df['FB']
y = daily_return_df['TWTR']
```

```python
daily_return_df.corr() #checking the correlations between stock prices
```

```python
# plotting the correlation between FB and twiter stock price
plt.figure(figsize = (15,10))
plt.scatter(x,y)
plt.xlabel('Facebook', color = 'red')
plt.ylabel('Twiter', color = 'red')
```

# 3. PLOT PIE CHART

```python
# Use matplotlib to plot a pie chart 
values  = [35, 25, 55, 15, 45]
color   = ['#e63946','#457b9d','#fcbf49','#fb8b24', '#84a98c']
labels  = ['Company A', 'Company B','Company C', 'Company D', 'Company E']
explode = [0,0.1,0,0,0]

plt.figure(figsize = (7,7))
plt.pie(values,colors = color,labels = labels, explode = explode, shadow = True,startangle = 60,autopct = '%1.1f%%')
plt.title('Stock Proportions', color = '#9e2a2b')
```

# 4. PLOT HISTOGRAMS

```python
bins = 40
data = daily_return_df['FB']
mean = daily_return_df['FB'].mean()
sdev = daily_return_df['FB'].std()
plt.figure(figsize=(15,10))
plt.hist(data,bins, color = '#0077b6');
plt.title(' Facebook Stock Price Distribution\n\n\n', color = '#023e8a')
plt.title('                      Mean = '+ str(mean) +'\nStandard Deviation = ' + str(sdev), loc = 'left', color = '#0077b6');
plt.xticks(color = 'red')
plt.yticks(color = 'red');

 
```

# 5. PLOT MULTIPLE PLOTS

```python
stock_df.plot('Date',['FB','NFLX'], figsize = (18,10), linewidth = 1, color= ['#e63946','#01497c'])
plt.ylabel('Price ($)', color = '#e63946')
plt.xlabel('')
plt.legend(prop={'size':16}, loc = 'upper left')
plt.title('FB & NFLX Stock Price Comparison', color = '#e63946');
```

# 6. PLOT SUBPLOTS

```python
plt.figure(figsize = (20,20))

plt.subplot(3,1,1) # 3 rows of plots, 1 column of plots, no. of the plot
plt.plot(stock_df['FB'], linewidth = 1, color = 'red')
plt.title('Facebook',color = 'red')

plt.subplot(3,1,2) # 3 rows of plots, 1 column of plots, no. of the plot
plt.plot(stock_df['NFLX'], linewidth = 1, color = 'Green')
plt.title('Netflix', color = 'Green')

plt.subplot(3,1,3) # 3 rows of plots, 1 column of plots, no. of the plot
plt.plot(stock_df['TWTR'], linewidth = 1, color = '#01497c')
plt.title('Twiter', color = '#01497c');
```

# 7. SEABRON SCATTERPLOT & COUNTPLOT

```python
# import breast cancer data
df_cancer = pd.read_csv('Breast_cancer.csv')
```

```python
# Check out the head of the dataframe
df_cancer.head()
```

```python
# Check out the shape of the dataframe
df_cancer.shape
```

```python
df_cancer['target'] = df_cancer['diagnosis'].apply(lambda x : 1 if x == 'B' else 0)

# Another way to do the above line of code and much faster in processing.

df_cancer['target2'] = np.where(df_cancer.diagnosis == 'B',1,0)
```

```python
# Plot scatter plot between mean area and mean smoothness
plt.figure(figsize = (10,10))

sns.scatterplot(data = df_cancer, x = 'area_mean', y = 'smoothness_mean', hue = 'diagnosis', palette = 'Accent');
```

```python
# Let's print out countplot to know how many samples belong to class #0 and #1
plt.figure(figsize = (10,10))
sns.countplot(x = df_cancer['diagnosis'], palette = 'Accent');
```

# 9. SEABORN PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS


```python
# Plot the pairplot
vars = ['smoothness_mean','area_mean','perimeter_mean','texture_mean','radius_mean']
sns.pairplot(df_cancer, hue = 'diagnosis', vars = vars );
```

```python
# Strong correlation between the mean radius and mean perimeter, mean area and mean primeter
corr_df = df_cancer[['smoothness_mean','area_mean','perimeter_mean','texture_mean','radius_mean']]
plt.figure(figsize = (30, 30)) 
sns.set(font_scale = 2.5) # make the font bigger

sns.heatmap(data = corr_df.corr(), annot = True, cmap = 'Blues'); 
```

```python
# Plot Histogram distripution for the men radius
plt.figure(figsize = (10,8))
mean = df_cancer['radius_mean'].mean()


sns.histplot(df_cancer['radius_mean'], bins = 25, kde = True)
plt.title('Mean = '+ str(mean));
```
